```python
# try manually recreating analysis
tfile = ROOT.TFile(file2, "read")
sumw = 160273396662272.0
lumi = 44307.4
# lumi = 139
xs = 14126.0
filter_eff = 0.00867775
kFactor = 1.0
reco_weight = f"(weight * {lumi * xs * filter_eff * kFactor}) / {sumw}"
mtw = "mt(TauPt, MET_met, TauPhi, MET_phi)"
th1_model = ["", "", 20, np.geomspace(170, 1000, 21)]

pass_presel = r"(passReco == 1) && (TauBaselineWP == 1) && (abs(TauCharge) == 1) && ((MatchedTruthParticle_isTau + MatchedTruthParticle_isElectron + MatchedTruthParticle_isMuon + MatchedTruthParticle_isPhoton) <= 1)"
pass_taupt170 = r"TauPt > 170"
pass_mtw150 = r"MTW > 150"
pass_loose = r"(TauBDTEleScore > 0.05) && ((TauRNNJetScore > 0.15) * (TauNCoreTracks == 1) + (TauRNNJetScore > 0.25) * (TauNCoreTracks == 3))"
pass_met150 = r"MET_met > 150"

df_sys_down = (  # sys_down
    ROOT.RDataFrame(prefix + check_sys_down, file2)
    .Define("MTW", mtw)
    .Redefine("TauPt", "TauPt / 1000")
    .Redefine("MTW", "MTW / 1000")
    .Redefine("MET_met", "MET_met / 1000")
    .Filter(pass_presel)  # pass_presel
    .Filter(pass_taupt170)
    # .Filter(pass_mtw150)
    .Filter(pass_loose)
    # .Filter(pass_met150)
    .Define("reco_weight", reco_weight)
)
df_nom = (  # nominal
    ROOT.RDataFrame(nominal, file2)
    .Define("MTW", mtw)
    .Redefine("TauPt", "TauPt / 1000")
    .Redefine("MTW", "MTW / 1000")
    .Redefine("MET_met", "MET_met / 1000")
    .Filter(pass_presel)  # pass_presel
    .Filter(pass_taupt170)
    # .Filter(pass_mtw150)
    .Filter(pass_loose)
    # .Filter(pass_met150)
    .Define("reco_weight", reco_weight)
)
df_sys_up = (  # sys_up
    ROOT.RDataFrame(prefix + check_sys_up, file2)
    .Define("MTW", mtw)
    .Redefine("TauPt", "TauPt / 1000")
    .Redefine("MTW", "MTW / 1000")
    .Redefine("MET_met", "MET_met / 1000")
    .Filter(pass_presel)  # pass_presel
    .Filter(pass_taupt170)
    # .Filter(pass_mtw150)
    .Filter(pass_loose)
    # .Filter(pass_met150)
    .Define("reco_weight", reco_weight)
)

# fill
hist_sys_up = ROOT.TH1F(*th1_model)
hist_sys_down = ROOT.TH1F(*th1_model)
hist_nom = ROOT.TH1F(*th1_model)
hist_sys_up = df_sys_up.Fill(hist_sys_up, ["TauPt", "reco_weight"]).GetValue()
hist_sys_down = df_sys_down.Fill(hist_sys_down, ["TauPt", "reco_weight"]).GetValue()
hist_nom = df_nom.Fill(hist_nom, ["TauPt", "reco_weight"]).GetValue()

print("sys_down: ", ROOT_utils.get_th1_bin_values(hist_sys_down))
print("nominal: ", ROOT_utils.get_th1_bin_values(hist_nom))
print("sys_up: ", ROOT_utils.get_th1_bin_values(hist_sys_up))
```
